### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

#1. Análise do faturamento


```python
faturamento_loja1 = loja1.iloc[:, 2].sum()
faturamento_loja2 = loja2.iloc[:, 2].sum()
faturamento_loja3 = loja3.iloc[:, 2].sum()
faturamento_loja4 = loja4.iloc[:, 2].sum()

print(f"Faturamento total da Loja 1: R${faturamento_loja1:.2f}")
print(f"Faturamento total da Loja 2: R${faturamento_loja2:.2f}")
print(f"Faturamento total da Loja 3: R${faturamento_loja3:.2f}")
print(f"Faturamento total da Loja 4: R${faturamento_loja4:.2f}")

```

# 2. Vendas por Categoria


```python
vendas_categoria_loja1_df = loja1['Categoria do Produto'].value_counts()
vendas_categoria_loja2_df = loja2['Categoria do Produto'].value_counts()
vendas_categoria_loja3_df = loja3['Categoria do Produto'].value_counts()
vendas_categoria_loja4_df = loja4['Categoria do Produto'].value_counts()

# Exibir os resultados
print("\nVENDAS POR CATEGORIA - LOJA 1:")
print(vendas_categoria_loja1_df)
print("\nVENDAS POR CATEGORIA - LOJA 2:")
print(vendas_categoria_loja2_df)
print("\nVENDAS POR CATEGORIA - LOJA 3:")
print(vendas_categoria_loja3_df)
print("\nVENDAS POR CATEGORIA - LOJA 4:")
print(vendas_categoria_loja4_df)
```

```python
# Loja 1
vendas_categoria_loja1 = loja1['Categoria do Produto'].value_counts().reset_index()
vendas_categoria_loja1.columns = ['Categorias', 'Quantidade de Vendas']
print("Vendas por categoria - Loja 1:")
print(vendas_categoria_loja1)
print("\n")

# Loja 2
vendas_categoria_loja2 = loja2['Categoria do Produto'].value_counts().reset_index()
vendas_categoria_loja2.columns = ['Categorias', 'Quantidade de Vendas']
print("Vendas por categoria - Loja 2:")
print(vendas_categoria_loja2)
print("\n")

# Loja 3
vendas_categoria_loja3 = loja3['Categoria do Produto'].value_counts().reset_index()
vendas_categoria_loja3.columns = ['Categorias', 'Quantidade de Vendas']
print("Vendas por categoria - Loja 3:")
print(vendas_categoria_loja3)
print("\n")

# Loja 4
vendas_categoria_loja4 = loja4['Categoria do Produto'].value_counts().reset_index()
vendas_categoria_loja4.columns = ['Categorias', 'Quantidade de Vendas']
print("Vendas por categoria - Loja 4:")
print(vendas_categoria_loja4)
print("\n")
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacao_loja1 = loja1['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f"Média de avaliação - Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de avaliação - Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de avaliação - Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de avaliação - Loja 4: {media_avaliacao_loja4:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
def mais_e_menos_vendidos(df, nome_loja):
    contagem = df['Produto'].value_counts()
    mais_vendido = contagem.idxmax()
    menos_vendido = contagem.idxmin()

    print(f" {nome_loja}")
    print(f" Produto mais vendido: {mais_vendido} ({contagem.max()} vendas)")
    print(f" Produto menos vendido: {menos_vendido} ({contagem.min()} vendas)")
    print("-" * 40)

mais_e_menos_vendidos(loja1, "Loja 1")
mais_e_menos_vendidos(loja2, "Loja 2")
mais_e_menos_vendidos(loja3, "Loja 3")
mais_e_menos_vendidos(loja4, "Loja 4")

```

# 5. Frete Médio por Loja


```python
# Cálculo do custo médio de frete por loja
frete_medio_loja1 = loja1['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

# Exibindo resultados formatados
print(f" Custo médio de frete - Loja 1: R$ {frete_medio_loja1:.2f}")
print(f" Custo médio de frete - Loja 2: R$ {frete_medio_loja2:.2f}")
print(f" Custo médio de frete - Loja 3: R$ {frete_medio_loja3:.2f}")
print(f" Custo médio de frete - Loja 4: R$ {frete_medio_loja4:.2f}")
```

# **1. Gráfico de barras – Faturamento total por loja**

```python
import matplotlib.pyplot as plt

faturamentos = [loja1['Preço'].sum(), loja2['Preço'].sum(), loja3['Preço'].sum(), loja4['Preço'].sum()]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8,5))
plt.bar(lojas, faturamentos, color='green')
plt.title('Faturamento Total por Loja')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# **2. Gráfico de pizza – Distribuição de vendas por categoria (ex: Loja 1)**

```python
categorias = loja1['Categoria do Produto'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(categorias, labels=categorias.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Vendas por Categoria - Loja 1\n')
plt.axis('equal')  # Deixa o gráfico redondinho
plt.show()
```

# **3. Gráfico de dispersão (scatter plot) – Avaliações vs. Preço (Loja 1)**

```python
top10_produtos = loja1['Produto'].value_counts().head(10).index

# Filtrando a loja para esses produtos e calculando a média de avaliação
avaliacao_media_top10 = loja1[loja1['Produto'].isin(top10_produtos)].groupby('Produto')['Avaliação da compra'].mean().sort_values()

# Plotando o gráfico
plt.figure(figsize=(10,6))
avaliacao_media_top10.plot(kind='barh', color='cornflowerblue')
plt.title('Avaliação Média dos 10 Produtos Mais Vendidos - Loja 1')
plt.xlabel('Avaliação Média')
plt.ylabel('Produto')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(loja1['Preço'], loja1['Avaliação da compra'], alpha=0.5, color='dodgerblue')
plt.title('Avaliação dos Produtos vs. Preço - Loja 1')
plt.xlabel('Preço (R$)')
plt.ylabel('Avaliação')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# **Gráfico: Produtos com Menor Avaliação Média (Loja 1)**

```python
# Agrupa os produtos e calcula a média de avaliação
avaliacao_media_por_produto = loja1.groupby('Produto')['Avaliação da compra'].mean()

# Seleciona os 10 com menor média
menores_avaliacoes = avaliacao_media_por_produto.sort_values().head(10)

# Plotando o gráfico
plt.figure(figsize=(10,6))
menores_avaliacoes.plot(kind='barh', color='salmon')
plt.title('10 Produtos com Menor Avaliação Média - Loja 1')
plt.xlabel('Avaliação Média')
plt.ylabel('Produto')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
